```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
df = pd.read_csv("C:\\Users\\Joe\\Desktop\\py\\bddNoSql\\Promoteur_imo.csv", delimiter=";")
df
```

```python
series = df["id_lot"]
```

```python
result = []
for ser in series:
    splited = ser.split('_')
    splited = list(filter(None, splited))
    nelem = len(splited)
    if nelem == 3:
        splited.insert(1,'inconnue')
    if nelem == 5:
        splited[1:3] = ['_'.join(splited[1:3])]
    result.append(splited)
result
```

```python
df3 = pd.DataFrame(result,columns=['surface','etage','nom_du_programme','promoteur'])
df3
```
